```python
def calculate_attendance(attended, total):
    if total > 0:
        return (attended / total) * 100
    else:
        return 0

def get_student_status(fees):
    if fees >= 7000:
        return "Premium"
    else:
        return "Standard"
        
def has_warning(attendance_percent):
    if attendance_percent < 75:
        return "Yes"
    else:
        return "No"
```

```python
def collect_student_data():
    name = input("Enter Student Name: ").strip()
    batch = input("Enter Batch Name: ").strip()

    fees = input("Enter Fees Paid: ")
    total_classes = input("Enter Total Classes: ")
    attended_classes = input("Enter Classes Attended: ")

    if not (fees.isdigit() and total_classes.isdigit() and attended_classes.isdigit()):
        print("Invalid numeric input. Record skipped.\n")
        return None

    fees = int(fees)
    total_classes = int(total_classes)
    attended_classes = int(attended_classes)

    if attended_classes > total_classes:
        print("Attended classes cannot exceed total classes. Record skipped.\n")
        return None

    student = {
        "name": name,
        "batch": batch,
        "fees": fees,
        "total": total_classes,
        "attended": attended_classes
    }

    return student
```

```python
def display_student_summary(student):
    attendance = calculate_attendance(student["attended"], student["total"])
    status = get_student_status(student["fees"])
    warning = has_warning(attendance)

    print("------ STUDENT SUMMARY ------")
    print("Name:", student["name"])
    print("Batch:", student["batch"])
    print("Attendance:", round(attendance, 2), "%")
    print("Status:", status)
    print("Warning:", warning)
    print()
```

```python
def generate_management_report(students):
    total_students = len(students)
    premium = 0
    standard = 0
    total_attendance = 0
    low_attendance_students = []

    for s in students:
        attendance = calculate_attendance(s["attended"], s["total"])
        total_attendance += attendance

        if s["fees"] >= 7000:
            premium += 1
        else:
            standard += 1

        if attendance < 75:
            low_attendance_students.append(s["name"])

    average_attendance = total_attendance / total_students if total_students > 0 else 0

    file = open("training_center_report.txt", "w")
    file.write("TRAINING CENTER MANAGEMENT REPORT\n")
    file.write("--------------------------------\n")
    file.write("Total Students: " + str(total_students) + "\n")
    file.write("Premium Students: " + str(premium) + "\n")
    file.write("Standard Students: " + str(standard) + "\n")
    file.write("Average Attendance: " + str(round(average_attendance, 2)) + "%\n")
    file.write("Low Attendance Students:\n")

    for name in low_attendance_students:
        file.write("- " + name + "\n")

    file.close()
```

```python
def run_system():
    students = []

    while True:
        print("1. Add Student")
        print("2. View All Students")
        print("3. Generate Management Report")
        print("4. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            student = collect_student_data()
            if student:
                students.append(student)
                display_student_summary(student)

        elif choice == "2":
            if not students:
                print("No Students Exist !")
            else:
                for s in students:
                    display_student_summary(s)

        elif choice == "3":
            generate_management_report(students)
            print("Report generated: training_center_report.txt\n")

        elif choice == "4":
            print("Exiting system.")
            break

        else:
            print("Invalid choice.\n")
```

```python
run_system()
```
