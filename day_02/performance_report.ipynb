```python
# Performance Report generation

import math
region_info = {}
def sales_input(region, sale):
    if region not in region_info:
        region_info[region] = []
        region_info[region].append(sale)
    else:
        region_info[region].append(sale)

def total_company_sales():
    total_sales = 0
    for region in region_info:
        total_sales += sum(region_info[region])
    return total_sales

def avg_sales(total_sales):
    return total_sales/len(region_info)

def best_worst_region():
    best_region = ""
    worst_region = ""

    maxi = -math.inf
    mini = math.inf
    
    for region in region_info:
        region_sum = sum(region_info[region])
        if region_sum > maxi:
            maxi = region_sum
            best_region = region
            
        if region_sum < mini:
            mini = region_sum
            worst_region = region
    return best_region, worst_region

def performance_report():
    with open("performance_report.txt", "w") as f:
        f.write("==== Performance Report ====\n")
        for region in region_info:
            f.write(region + ": " + str(sum(region_info[region])))
            f.write("\n")
        f.write("\nTotal Sale: " + str(total_company_sales()))
        total = total_company_sales()
        f.write("\nAverage Sale: " + str(avg_sales(total)))

        best_reg, worst_reg = best_worst_region()
        f.write(f"\nBest Region: {best_reg}")
        f.write(f"\nWorst Region: {worst_reg}")


def analytics_summary():
    print("=== Region Wise Company Sales ====")
    for region in region_info:
        print(region, ":", sum(region_info[region]))
    print()
    print("=== Total Sales ===")
    print("Total Sales:", total_company_sales())
    print()
    print("=== Average Sale ===")
    total = total_company_sales()
    print("Average Sale:", avg_sales(total))
    print()
    print("=== Best and Worst Performing Region ===")
    best_reg, worst_reg = best_worst_region()
    print("Best Region:", best_reg)
    print("Worst Region:", worst_reg)
    print()
```

```python
sales_input('Bangalore', 5000)
sales_input('Bangalore', 5990)
sales_input('Bangalore', 8000)

sales_input('Chennai', 6000)
sales_input('Chennai', 6900)
sales_input('Chennai', 7000)

sales_input('Hyderabad', 8000)
sales_input('Hyderabad', 2000)
sales_input('Hyderabad', 9000)

sales_input('Mumbai', 9900)
sales_input('Mumbai', 1900)
sales_input('Mumbai', 9100)

sales_input('Delhi', 5000)
sales_input('Delhi', 3000)
sales_input('Delhi', 5700)
```

```python
analytics_summary()
```

```python
try:
    performance_report()
    print('Performance Report Saved !')
except Exception as e:
    print("Exception generated: ", e)
```
